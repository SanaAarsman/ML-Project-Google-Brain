## Univariate Analysis

```python
import matplotlib.pyplot as plt
```

```python
# Selecting the first 17 columns

first_17 = df.iloc[:, :17]
first_17.columns
```

```python
# Splitting Numerical from Categorical

num_cols = first_17.select_dtypes(include="number").columns
cat_cols = first_17.select_dtypes(include="object").columns
```

```python
first_17[num_cols].describe()
```

### Histogram for continuous variables

```python

cont_cols = ["Age", "DailyRate", "DistanceFromHome", "HourlyRate"]

for col in cont_cols:
    plt.hist(df[col], bins=20)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

* Most workers are between 30 - 40 year old
* Distance from home shows most people live near work
* Hourly & Daily Rate have a fair distribution

### Boxplot for Outliers

```python
for col in cont_cols:
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
```

### Barplots for categorical 

```python
ord_cols = [
    "Education",
    "EnvironmentSatisfaction",
    "JobInvolvement",
    "JobLevel",
    "JobSatisfaction"
]

for col in ord_cols:
    df[col].value_counts().sort_index().plot(kind="bar")
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

* Most values are well distributed, with a few exceptions like JobLevel having the most values in level 1 & 2, JobInvolvement and Education in 3.

### Attrition Barplot

```python
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Attrition Distribution")
plt.show()
```
